## Jawaban soal no.1

Menentukan jumlah plat genap dan ganjil

```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']
```

```python
# Function to define odd and even plate
def check(x):
    x = int(x.split()[1]) # Get numerical values from 'number plate'
    result = 'genap' if x % 2 == 0 else 'ganjil'
    return result;
```

```python
# Initial values
even = 0
odd = 0

# Iteration to counting even and odd number plate
for x in list_plat_mobil:
    # plat = x
    if check(x) == 'genap': # Conditional statement
        even += 1
    else:
        odd += 1
  # print(f'plat {plat} adalah {cek(x)}')

print(f'Jumlah plat Ganjil adalah {odd} dan Genap adalah {even}')
```

## Jawaban soal no.2

Deret Fibbonanci

```python
# Function fibbonanci from (2, 3)
def fibonacci(n):
    sequence = [2, 3] # Initial values
    while len(sequence) < n:
        next_value = sequence[len(sequence) - 1] + sequence[len(sequence) - 2]
        sequence.append(next_value)
    return sequence;
```

```python
print("12 angka pertama dalam deret Fibonacci:")
print(fibonacci(12))
```

function *string_processor*

```python
# Function string processor
def string_processor(x, process):
    process = process.lower()
    x = str(x)
    x = x.lower()
    if process == 'first_char':
      rsl = x[0]
    elif process == 'last_char':
      rsl = x[-1]
    elif process == 'num_chars':
      rsl = len(x)
    elif process == 'num_vowels':
      rsl = num_vowels(x)
    return rsl;

# Define vowel
def num_vowels(x):
  count = 0
  vowels = ["a", "e", "i", "o", "u"]
  for i in range(len(x)):
    if x[i] in vowels:
        count += 1
  return count;
```

## Jawaban soal no.7

Kelompok **segment** (high, medium, dan low) berdasarkan performance Sales

```python
df = pd.read_excel('/content/drive/MyDrive/0. Rakamin/superstore.xls')
print(df.duplicated().any())

try:
  df['Year'] = df['Order Date'].apply(lambda x: x.year) # Create cloumn Year
except:
  print("Create column Year failed: check format 'Order date'")

# Aggregation Sales and Profit by year and user
df_n = df.groupby(['Year', 'Customer ID', 'Customer Name', 'Segment']).agg({'Sales' : 'sum'}).reset_index()

df_n.columns = ['Year', 'Customer ID', 'Customer Name', 'Segment', 'Sales_total']

df_n.head()
```

```python
# Show Descriptive statistics on Sales_total by Year

df_o = df_n.groupby('Year')['Sales_total'].describe()

percentiles = [60, 70, 80, 90, 95, 99]
for p in percentiles:
    df_o[f'{p}%'] = df_n.groupby('Year')['Sales_total'].apply(lambda x: np.percentile(x, p))

# Calculate mean + std value which is represented 1 sigma significance level
df_o['mean_std'] = df_o['mean'] + df_o['std']
df_o = df_o.reset_index()

df_o
```

```python
# Merging dataframe df_o and df_n to show quantile by User

df_n = df_n[['Year', 'Customer ID', 'Sales_total']]

df_o = df_o[['Year', 'mean', 'std', 'mean_std', '75%', '90%', '95%']]

df_m = df_n.merge(df_o, on='Year')

df_m.sample(10)
```

Alasan user dibagi berdasarkan persentil yaitu dengan melihat kemampuan sales berdasarkan statistik deskriptif dan jumlah user per segment untuk setiap tahun. 

Berdasarkan data dapat dilihat dari ukuran pemusatan data bahwa user dapat mencapai sales > 2000 pada persentil ~90% dan sales > 900 pada persentil ~75%. Nilai persentil ~75% lebih tinggi daripada nilai rerata sales per user pada setiap tahun (nilai rerata sales berkisar 800 - 1000). Sehingga nilai **persentil ~75%** kami pilih sebagai **nilai batas atas segment low**, hal ini dilakukan untuk mengambil data user yang memiliki sales diatas rerata saja pada segment medium dan high. 

**Nilai batas bawah high** diambil dari **nilai rerata ditambah dengan simpangan baku sales** yang juga secara statistik disebut **memiliki signifikasi 1σ**. Nilai 1σ untuk setiap tahun > 1900. User yang memiliki sales **diatas 1σ** dianggap sebagai user yang memiliki sales cukup tinggi jika dibandingkan dengan user secara umum. Nilai ini dipilih berdasarkan distribusi normal. User yang memiliki nilai 1σ berdasarkan statistik hanya 15.87% dari populasi user.

(Ref: https://thecuriousastronomer.wordpress.com/2014/06/26/what-does-a-1-sigma-3-sigma-or-5-sigma-detection-mean/)

```python
# Function for get user segment by sales

def get_segment(sales):
  if sales['Sales_total'] < sales['75%']:
    return 'Low'
  elif sales['Sales_total'] < sales['mean_std']:
    return 'Medium'
  else:
    return 'High'

df_m['User_segment'] = df_m.apply(lambda x: get_segment(x), axis = 1)

df_m.tail(10)
```

```python
# Count total user segment by unique and count cust ID

df_fin = df_m.groupby(['Year','User_segment']).agg({'Customer ID' : 'nunique'}).reset_index()

df_fin.head(10)
```

```python
# Count user segment in 2017

rsl = df_fin[df_fin['Year'] == 2017].reset_index(drop = True)
rsl
```

```python
print(f"Jumlah customer pada tahun 2017 dengan segment low {rsl.loc[0]['Customer ID']} customer, medium {rsl.loc[1]['Customer ID']} cutomer, dan high {rsl.loc[2]['Customer ID']} customer.")
```
