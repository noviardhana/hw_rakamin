### 3. Mengolah Dataset (36pts)

Dengan menggunakan dataset `application_processed.csv` buatlah dataframe baru dengan ketentuan:
1. Ambil semua users yang memiliki `email`. (ditandai dengan nilai pada column `flag_email` = 1) dan tidak memilik data `duplicates`. (6pts)
2. Buatlah column baru bernama `has_car_and_property` dimana nilainya merupakan range dari `0-1`. `1` apabila `flag_own_car` bernilai `1` dan `flag_own_property` bernilai `1`. `0` apabila terdapat atau semua nilai pada column `flag_own_car` dan `flag_own_property` terdapat angka 0. (6pts)
3. Ambil column yang diperlukan yaitu `id`, `gender`, `has_car_and_property`, `income_type`, `education_type`, `family_status`, `housing_type`, `occupation_type`, `amount_income_total`. (6pts)
4. Lakukan filter dengan ketentuan sebagai berikut: (6pts)
    - Memiliki `car` dan `property`
    - Tidak ada nilai null pada column `occupation_type`
    - Memiliki value pada column `occupation_type` dengan pola `staff`.
5. Sudah dilakukan sorting berdasarkan `income_type` dari `A-Z` dan `amount_income_total` dari `besar ke kecil`. (6pts)
6. Simpan kedalam bentuk `.csv` dengan nama `homework_3_result.csv`. (6pts)

**Expected Result:**

![Expected Result](assets/homework_exercise_3_result.png)

```python
df_application_filtered = df_application[df_application["flag_email"] == 1]
df_application_filtered = df_application_filtered.drop_duplicates()
df_application_filtered["has_car_and_property"] = df_application_filtered["flag_own_car"] * df_application_filtered["flag_own_property"]
df_application_filtered = df_application_filtered[[
    "id", "gender", "has_car_and_property", "income_type", "education_type", "family_status",
    "housing_type", "occupation_type", "amount_income_total"
]]
df_application_filtered = df_application_filtered[
    (df_application_filtered["has_car_and_property"] == 1)
    & (df_application_filtered["occupation_type"].astype(str).str.contains("staff"))
    & (df_application_filtered["occupation_type"].notnull())
]
df_application_sorted = df_application_filtered.sort_values(
    by=["income_type", "amount_income_total"],
    ascending=[True, False]
)
df_application_sorted.to_csv("homework_3_result.csv", index=None)
pd.read_csv("homework_3_result.csv")
```

### 5. Memanfaatkan Pivot Tabel (15pts)

Buatlah dataframe baru menggunakan fungsi pivot table untuk melihat `Embarked` apa saja yang `average_female_ticket_price` lebih besar daripada `average_male_ticket_price`. Informasi `average_female_ticket_price` dan `average_male_ticket_price` didapatkan dari rata-rata pada column `Fare`.

**Hint:**
1. menggunakan pivot table
2. dapat dilihat pada topik `DataFrame Aggregation` bagian `Pivot Table | Reshape Columns and Rows`

**Expected Result:**

![Expected Result](assets/homework_exercise_5_result.png)

```python
df_titanic = pd.read_csv("titanic.csv")
df_analysis_titanic = pd.pivot_table(
    df_titanic,
    index=["Embarked"],
    values=["Fare"],
    columns=["Sex"],
    aggfunc={
        "Fare": "mean"
    }
)
df_analysis_titanic = df_analysis_titanic.reset_index()
df_analysis_titanic.columns = ["Embarked", "avg_female_ticket_price", "avg_male_ticket_price"]
df_analysis_titanic = df_analysis_titanic[df_analysis_titanic["avg_female_ticket_price"] >= df_analysis_titanic["avg_male_ticket_price"]]
df_analysis_titanic
```

### 7. Merge DataFrame (25pts)

Buatlah 2 dataframe mengikuti petunjuk berikut:

df_1 :  Filter dataset 01 telecom_revenue.csv dengan multiple kondisi (5pts)
- MonthlyRevenue > 10 
- Occupation terdiri dari Professional, Student, and Crafts

df_2 :  Filter dataset 02 telecom_usage.csv dengan multiple kondisi (5pt)
- nilai UnansweredCalls > .BlockedCalls

Jawablah pertanyaan berikut ini:
1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

**Expected Result:**

![Expected Result](assets/homework_exercise_7_result.png)

```python
df1 = pd.read_csv("01 telecom_revenue.csv")
df1 = df1[
    (df1["MonthlyRevenue"] > 10)
    & (df1["Occupation"].isin(["Professional", "Students", "Crafts"]))
]

df2 = pd.read_csv("02 telecom_usage.csv")
df2 = df2[
    (df2["UnansweredCalls"] > df2["BlockedCalls"])
]

df_merge = df1.merge(df2, on="CustomerID")
df_merge = df_merge.groupby("Occupation").agg({
    "CustomerID": "count",
    "DroppedCalls": "mean"
})
df_merge
```
